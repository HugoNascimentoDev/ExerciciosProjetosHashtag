Exercício 1: Validação de Investimento (Setor Financeiro) Uma corretora de valores quer automatizar a recomendação básica de perfil. Crie um programa que peça ao usuário o valor que ele deseja investir.
1.	Se o valor for menor que R$ 1.000,00, exiba: "Perfil iniciante: Sugerimos Tesouro Direto".
2.	Se o valor for entre R$ 1.000,00 e R$ 5.000,00 (inclusive), exiba: "Perfil moderado: Sugerimos Fundos Imobiliários".
3.	Se o valor for acima de R$ 5.000,00, exiba: "Perfil arrojado: Sugerimos Ações". *Lembre-se de tratar o input caso o usuário digite "R$" ou use vírgula.*


```python
valor = float(input('Qual o valor que deseja investir? R$ '))
tipo_perfil = ''
if valor < 1000 and valor > 0:
    tipo_perfil = 'Iniciante: Sugerimos Tesouro Direto'
elif valor >= 1000 and valor < 5000:
    tipo_perfil = 'Moderado: Sugerimos Fundos Imobiliários'
elif valor >= 5000:
    tipo_perfil = 'Arrojado: Sugerimos Ações'
else:
    print('Valor Inválido! Tente Novamente! ')

if valor > 0:
    print(f'Perfil {tipo_perfil}')

```

Exercício 2: Controle de Acesso ao Sistema (Setor de Segurança) Você tem uma lista de e-mails de administradores: admins = ["ana@empresa.com", "guilherme@empresa.com", "felipe@empresa.com"]. Crie um programa que peça o e-mail do usuário. O programa deve:
1.	Padronizar o e-mail (letras minúsculas e sem espaços).
2.	Verificar se o e-mail está na lista de admins.
3.	Se estiver, exibir: "Acesso liberado! Bem-vindo ao painel de controle".
4.	Caso contrário, exibir: "Acesso negado. Você não tem permissões de administrador".


```python
admins = ["ana@empresa.com", "guilherme@empresa.com", "felipe@empresa.com"]

email_usuario = input('Digite o seu email: ').lower().strip()

if email_usuario in admins:
    print('Acesso Liberado!!! Bem-vindo ao painel de controle!')
else: 
    print('Acesso Negado! Você não tem ermissões de administrador!')

```

Exercício 3: Cálculo de Desconto Progressivo (Setor de Vendas) Um e-commerce aplica descontos automáticos no carrinho. Crie um programa que receba o valor total da compra e aplique a seguinte lógica:
●	Compras a partir de R$ 500,00: 15% de desconto.
●	Compras a partir de R$ 200,00 (e menos de 500): 10% de desconto.
●	Compras abaixo de R$ 200,00: Sem desconto. O programa deve exibir o valor do desconto e o valor final a pagar, formatados em R$.


```python
valor_compra = float(input('Digite o valor total de compra: R$ '))

taxa_desconto = 0

if valor_compra >= 500:
    taxa_desconto = 0.15
elif valor_compra >= 200:
    taxa_desconto = 0.10
else:
    taxa_desconto = 0

valor_desconto =  (valor_compra * taxa_desconto)
valor_liquido = valor_compra - valor_desconto

print(f'''Valor Total: R$ {valor_compra:.2f}
Taxa desconto: {taxa_desconto:.2%}
Valor desconto: R$ {valor_desconto:.2f}
Valor Líquido: R$ {valor_liquido:.2f}''')
```

Exercício 4: Análise de Metas Combinadas (Setor Comercial) Uma empresa paga bônus se a meta individual do vendedor E a meta da loja forem batidas.
1.	Peça as vendas do vendedor e a meta individual dele.
2.	Peça as vendas totais da loja e a meta da loja.
3.	Se o vendedor bater a meta dele E a loja bater a meta total, o bônus é de 20% sobre as vendas do vendedor.
4.	Caso contrário, o bônus é zero. Exiba a mensagem: "Seu bônus este mês é de: R$[valor]".


```python
vendas_vendedor = int(input('Digite o valor das suas vendas: '))
meta_vendedor = int(input('Digite o valor da meta vendedor: '))

vendas_loja = int(input('Digite o valor total da venda loja: '))
meta_loja = int(input('Digite o valor da meta da loja: '))

taxa_bonus = 0

if vendas_vendedor >= meta_vendedor and vendas_loja >= meta_loja:
    taxa_bonus = 0.20
else:
    taxa_bonus = 0

valor_bonus =  (vendas_vendedor * taxa_bonus)

print(f'Seu bônus este mês é de R$ {valor_bonus:.2f}')
```

Exercício 5: Sistema de Triagem de E-mails (Setor de Customer Experience) Crie um sistema que ajude a filtrar para qual departamento uma reclamação deve ir. O usuário deve digitar o assunto do e-mail.
●	Se no assunto aparecer a palavra "pagamento" ou "boleto", exiba: "Encaminhado para o Financeiro".
●	Se no assunto aparecer a palavra "entrega" ou "atraso", exiba: "Encaminhado para a Logística".


```python
assunto_email = input('Digite o assunto do email: ').strip().lower()

if 'pagamento' in assunto_email or 'boleto' in assunto_email:
    print('Email encaminhado para o financeiro!')
elif 'entrega' in assunto_email or 'atraso' in assunto_email:
    print('Email encaminhado para a Logística!')
else:
    print('Não conseguimos identificar o departamento responsável! Estaremos encaminhando diretamente ao time de Customer Service!')
```
