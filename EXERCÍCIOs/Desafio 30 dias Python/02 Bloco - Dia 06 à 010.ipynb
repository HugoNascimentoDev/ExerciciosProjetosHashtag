Dia 6 – Média como critério
Enunciado:
Crie uma lista apenas com valores acima da média.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados:
[500, 1200, 800, 2000, 1500, 700]


```python
vendas = [500, 1200, 800, 2000, 1500, 700]

def verificando_vendas (lista):
    acima_da_media = []
    total_vendas = sum(lista)
    qtd_vendas = len(lista)
    media = total_vendas / qtd_vendas
    qtd_acima_media = 0
    
    for venda in lista:
        if venda > media:
            acima_da_media.append(venda)
            qtd_acima_media += 1
    perc_acima_media = (qtd_acima_media / qtd_vendas)

    return acima_da_media, media, perc_acima_media, qtd_acima_media, total_vendas, qtd_vendas

acima_da_media, media, perc_acima_media, qtd_acima_media, total_vendas, qtd_vendas = (verificando_vendas(vendas))

print(f'''Valor total de vendas: {total_vendas}
Quantidade total de vendas: {qtd_vendas}
Média de vendas: {media:.2f}
Quantidade de vendas acima da média: {qtd_acima_media}
Percentual acima da média: {perc_acima_media:.2%}''')
for venda in acima_da_media:
    print(f'Valor da Venda acima da média: {venda}')

```
