Dia 6 – Média como critério
Enunciado:
Crie uma lista apenas com valores acima da média.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados:
[500, 1200, 800, 2000, 1500, 700]


```python
vendas = [500, 1200, 800, 2000, 1500, 700]

def verificando_vendas (lista):
    acima_da_media = []
    total_vendas = sum(lista)
    qtd_vendas = len(lista)
    media = total_vendas / qtd_vendas
    qtd_acima_media = 0
    
    for venda in lista:
        if venda > media:
            acima_da_media.append(venda)
            qtd_acima_media += 1
    perc_acima_media = (qtd_acima_media / qtd_vendas)

    return acima_da_media, media, perc_acima_media, qtd_acima_media, total_vendas, qtd_vendas

acima_da_media, media, perc_acima_media, qtd_acima_media, total_vendas, qtd_vendas = (verificando_vendas(vendas))

print(f'''Valor total de vendas: {total_vendas}
Quantidade total de vendas: {qtd_vendas}
Média de vendas: {media:.2f}
Quantidade de vendas acima da média: {qtd_acima_media}
Percentual acima da média: {perc_acima_media:.2%}''')
for venda in acima_da_media:
    print(f'Valor da Venda acima da média: {venda}')

```

Dia 7 – Ranking simples
Enunciado:
Gere um ranking de vendedores do maior para o menor faturamento.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados:
{'Ana': 35000, 'Bruno': 42000, 'Carlos': 28000, 'Daniela': 51000}


```python
vendas = {'Ana': 35000, 'Bruno': 42000, 'Carlos': 28000, 'Daniela': 51000}

def ordenador_vendas (dicionario):
    ordenado_menor_para_maior_dict = dict(sorted(dicionario.items(), key = lambda item: item[1]))
    ordenado_maior_para_menor_dict = dict(sorted(dicionario.items(), key = lambda item: item[1], reverse= True))

    return ordenado_menor_para_maior_dict, ordenado_maior_para_menor_dict

ordenado_menor_para_maior_dict, ordenado_maior_para_menor_dict = (ordenador_vendas(vendas))

print('=' * 50)
print('RANKING VENDAS (MENOR PARA O MAIOR)')
print('=' * 50)
contador = 0
for vendedor, t_vendas in ordenado_menor_para_maior_dict.items():
    contador += 1 
    print(f'{contador}º - {vendedor.upper()}: {t_vendas}')

print('=' * 50)
print('RANKING VENDAS (MAIOR PARA O MENOR)')
print('=' * 50)
contador = 0
for vendedores, to_vendas in ordenado_maior_para_menor_dict.items():
    contador += 1
    print(f'{contador}º - {vendedores.upper()}: {to_vendas}')

print('=' * 50)
```
