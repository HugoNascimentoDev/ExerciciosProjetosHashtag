# 📊 Dia 11 – Lista de Tuplas: Cálculo de Faturamento

## 🎯 Contexto

Você trabalha no setor comercial de uma empresa do segmento de baterias automotivas.  
Ao final do dia, recebeu um relatório simplificado contendo os valores faturados por produto.

Essas informações foram organizadas em uma **lista de tuplas**, onde:

- O primeiro elemento da tupla representa o nome do produto
- O segundo elemento representa o valor faturado com aquele produto

Sua missão é analisar esses dados e gerar indicadores básicos de faturamento.

---

## 🗂 Base de Dados

```python
[('Bateria 60Ah', 420), 
 ('Bateria 70Ah', 510), 
 ('Bateria 90Ah', 690)]

````python
lista_vendas = [('Bateria 60Ah', 420), 
                ('Bateria 70Ah', 510), 
                ('Bateria 90Ah', 690)]

faturamento_total = 0
prod_maior_faturamento = ''
qtd_maior_faturamento = 0
prod_menor_faturamento = ''
qtd_menor_faturamento = 0 

for i, venda in enumerate(lista_vendas):
    faturamento_total += venda[1]
    print(f'Produto: {venda[0].upper()} - Faturamento: {venda[1]}')
    if i == 0:
        prod_maior_faturamento = venda[0]
        qtd_maior_faturamento = venda[1]
        prod_menor_faturamento = venda[0]
        qtd_menor_faturamento = venda[1]
    if i > 0:
        if qtd_maior_faturamento < venda[1]:
            prod_maior_faturamento = venda[0]
            qtd_maior_faturamento = venda[1]
        if qtd_menor_faturamento > venda[1]:
            prod_menor_faturamento = venda[0]
            qtd_menor_faturamento = venda[1]



media_faturamento = faturamento_total / len(lista_vendas)  

print(f'''Faturamento Total: \t{faturamento_total}
Média de faturamento: \t{media_faturamento:.0f}
Item com MAIS vendas: {prod_maior_faturamento.upper()} - Faturamento: {qtd_maior_faturamento}
Item com MENOS vendas: {prod_menor_faturamento.upper()} - Faturamento: {qtd_menor_faturamento}''')
````
