# 📊 Dia 11 – Lista de Tuplas: Cálculo de Faturamento

## 🎯 Contexto

Você trabalha no setor comercial de uma empresa do segmento de baterias automotivas.  
Ao final do dia, recebeu um relatório simplificado contendo os valores faturados por produto.

Essas informações foram organizadas em uma **lista de tuplas**, onde:

- O primeiro elemento da tupla representa o nome do produto
- O segundo elemento representa o valor faturado com aquele produto

Sua missão é analisar esses dados e gerar indicadores básicos de faturamento.

---

## 🗂 Base de Dados

```python
[('Bateria 60Ah', 420), 
 ('Bateria 70Ah', 510), 
 ('Bateria 90Ah', 690)]

````python
lista_vendas = [('Bateria 60Ah', 420), 
                ('Bateria 70Ah', 510), 
                ('Bateria 90Ah', 690)]

faturamento_total = 0
prod_maior_faturamento = ''
qtd_maior_faturamento = 0
prod_menor_faturamento = ''
qtd_menor_faturamento = 0 

for i, venda in enumerate(lista_vendas):
    faturamento_total += venda[1]
    print(f'Produto: {venda[0].upper()} - Faturamento: {venda[1]}')
    if i == 0:
        prod_maior_faturamento = venda[0]
        qtd_maior_faturamento = venda[1]
        prod_menor_faturamento = venda[0]
        qtd_menor_faturamento = venda[1]
    if i > 0:
        if qtd_maior_faturamento < venda[1]:
            prod_maior_faturamento = venda[0]
            qtd_maior_faturamento = venda[1]
        if qtd_menor_faturamento > venda[1]:
            prod_menor_faturamento = venda[0]
            qtd_menor_faturamento = venda[1]



media_faturamento = faturamento_total / len(lista_vendas)  

print(f'''Faturamento Total: \t{faturamento_total}
Média de faturamento: \t{media_faturamento:.0f}
Item com MAIS vendas: {prod_maior_faturamento.upper()} - Faturamento: {qtd_maior_faturamento}
Item com MENOS vendas: {prod_menor_faturamento.upper()} - Faturamento: {qtd_menor_faturamento}''')
````

# 🚢 Dia 12 – Mini Desafio Autoral  
## 📦 Análise de Fretes Internacionais

### 🎯 Contexto

Você é Analista de Comércio Exterior em uma multinacional do setor de baterias automotivas.

No fechamento do mês, você recebeu um relatório simplificado contendo os valores de frete pagos por embarque.

Cada registro contém:
- Número do processo
- Cliente
- Modalidade (Marítimo ou Aéreo)
- Valor do frete (USD)

Sua missão é analisar esses dados e gerar indicadores para auxiliar na tomada de decisão.

---

## 🗂 Base de Dados

```python
fretes = [
    ('EXP001', 'Cliente A', 'Marítimo', 3200),
    ('EXP002', 'Cliente B', 'Aéreo', 5400),
    ('EXP003', 'Cliente A', 'Marítimo', 2800),
    ('EXP004', 'Cliente C', 'Aéreo', 6100),
    ('EXP005', 'Cliente B', 'Marítimo', 3500)
]

O que deve ser feito ? 

Você deverá:

Calcular o custo total de fretes no período.

Calcular:

Total gasto com fretes marítimos

Total gasto com fretes aéreos

Identificar:

O embarque com maior custo

O embarque com menor custo

Calcular a média de custo por embarque.

Calcular quanto cada cliente representa (%) do total de fretes.

Exibir um ranking dos embarques do maior para o menor valor.

````python
fretes = [
    ('EXP001', 'Cliente A', 'Marítimo', 3200),
    ('EXP002', 'Cliente B', 'Aéreo', 5400),
    ('EXP003', 'Cliente A', 'Marítimo', 2800),
    ('EXP004', 'Cliente C', 'Aéreo', 6100),
    ('EXP005', 'Cliente B', 'Marítimo', 3500)
]

custo_total_frete = 0
custo_total_aereo = 0
custo_total_maritimo = 0

for processo, cliente, modal, frete in fretes:
    custo_total_frete += frete
    if modal == 'Marítimo':
        custo_total_maritimo += frete
    else:
        custo_total_aereo += frete

maior_frete = max(fretes, key= lambda item: item[3])
menor_frete = min(fretes, key= lambda item: item[3])

custo_medio = custo_total_frete / len(fretes)

print('-'*50)
print(f'{'RESUMO DE DADOS':^50}')
print('-'*50)

print(f'''Custo Total de Fretes: {custo_total_frete}
Custo Médio por Embarque: {custo_medio:.0f}
Custo Total de Fretes Marítimos: {custo_total_maritimo}
Custo Total de Fretes Aéreos: {custo_total_aereo}
Maior Custo de Frete: {maior_frete[3]} - Processo: {maior_frete[0]} - Cliente: {maior_frete[1]} - Modal: {maior_frete[2]} 
Menor Custo de Frete: {menor_frete[3]} - Processo: {menor_frete[0]} - Cliente: {menor_frete[1]} - Modal: {menor_frete[2]}''')

fretes_ordenados_custo = sorted(fretes, key= lambda custo: custo[3], reverse= True)

print('-'*50)
print(f'{'RANKING DE CUSTOS POR EMBARQUE':^50}')
print('-'*50)

ranking_custo = 1

for processo, cliente, modal, frete in fretes_ordenados_custo:
    print(f'{ranking_custo}º - {processo} - {cliente} - {modal} - {frete}')
    ranking_custo += 1

custo_por_clientes = {}

for processo, cliente, modal, frete in fretes:
    if cliente not in custo_por_clientes:
        custo_por_clientes[cliente] = 0
    custo_por_clientes[cliente] += frete

print('-'*50)
print(f'{'RANKING DE CUSTOS POR CLIENTE':^50}')
print('-'*50)

ranking_cliente = 1

custo_por_clientes = {}

for processo, cliente, modal, frete in fretes:
    if cliente not in custo_por_clientes:
        custo_por_clientes[cliente] = 0
    custo_por_clientes[cliente] += frete

ranking_clientes = sorted(custo_por_clientes.items(), key = lambda item: item[1], reverse= True)

for cliente, custo in ranking_clientes:
    perc = custo / custo_total_frete
    print(f'{ranking_cliente}º - Cliente: {cliente} - Custo: {custo} - {perc:.2%} do total') 
    ranking_cliente += 1

print('-'*50)
print(f'{'RANKING DE CUSTOS POR MODAL':^50}')
print('-'*50)

ranking_modal = 1

custo_por_modal = {}

for processo, cliente, modal, frete in fretes:
    if modal not in custo_por_modal:
        custo_por_modal[modal] = 0
    custo_por_modal[modal] += frete

ranking_por_modal = sorted(custo_por_modal.items(), key = lambda item: item[1], reverse= True)

for modal, custo in ranking_por_modal:
    perc = custo / custo_total_frete
    print(f'{ranking_modal}º - Modal: {modal} \t- Custo: {custo} \t- {perc:.2%} do total') 
    ranking_modal += 1



````
