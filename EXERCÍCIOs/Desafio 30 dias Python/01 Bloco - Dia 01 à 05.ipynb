Dia 01:
Calcule o total e a média de vendas do dia.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: [1200, 850, 430, 2100, 980, 1500, 760]


```python
vendas = [1200, 850, 430, 2100, 980, 1500, 760]

print(vendas)

total = sum(vendas)
media = total / len(vendas)
maior_venda = max(vendas)
menor_venda = min(vendas)

print(f'O TOTAL de Vendas é: {total}.')
print(f'A MÉDIA de vendas é {media:.2f} por registro.')
print(f'A MAIOR venda foi: {maior_venda}.')
print(f'A MENOR venda foi: {menor_venda}.')
```

Dia 2 – Condições e contagem
Enunciado:
Conte quantos produtos possuem estoque maior que 10 e quantos estão abaixo de 5.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: [3, 15, 8, 22, 4, 12, 30]


```python
estoque = [3, 15, 8, 22, 4, 12, 30]

maior_dez = []
menor_cinco = []
estoque_seguro = []

for e in estoque:
    if e > 10:
        maior_dez.append(e)
    elif e < 5:
        menor_cinco.append(e)
    else:
        estoque_seguro.append(e)

print(f'''Total de produtos: {len(estoque)}
Total de produtos com estoque acima de 10 unidades: {len(maior_dez)}
Estoque acima de 10 unidades: {maior_dez}
Total de produtos com estoque abaixo de 5 unidades: {len(menor_cinco)}
Estoque abaixo de 5 unidades: {menor_cinco}
Estoque seguro: {estoque_seguro}''')
```

Dia 3 – Dicionários (estrutura chave/valor)
Enunciado:
Calcule o faturamento total de cada filial.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: {'Matriz': [12000, 15000, 11000], 'Filial Sul': [8000, 7200, 6500], 'Filial Norte': [9000, 8800]}

loja = {'Matriz': [12000, 15000, 11000], 'Filial Sul': [8000, 7200, 6500], 'Filial Norte': [9000, 8800]}




```python
loja = {'Matriz': [12000, 15000, 11000], 'Filial Sul': [8000, 7200, 6500], 'Filial Norte': [9000, 8800]}

total_de_vendas = 0
faturamento_total_por_loja = {}

for nome_loja, vendas in loja.items():  
    total_vendas_loja = sum(vendas)
    print(f'Loja: {nome_loja.upper()} - Vendas: {total_vendas_loja}' )
    total_de_vendas += total_vendas_loja
    faturamento_total_por_loja[nome_loja] = total_vendas_loja
print(f'''Total de vendas: {total_de_vendas}
Média de vendas: {total_de_vendas / len(loja):.2f}''')

print(faturamento_total_por_loja)


```

Dia 4 – Funções
Enunciado:
Crie uma função que receba uma lista de valores e retorne o total e a média.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados:
[1200, 850, 430, 2100, 980]



```python
lista_vendas = [1200, 850, 430, 2100, 980]

def calculando (lista):
    total = sum(lista)
    media = total / len(lista)   

    return total, media

total, media = calculando(lista_vendas)

print(f'Total de vendas: {total:.2f} - Média de vendas: {media:.2f}')



```

Enunciado:
Calcule o faturamento total de cada filial.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: {'Matriz': [12000, 15000, 11000], 'Filial Sul': [8000, 7200, 6500], 'Filial Norte': [9000, 8800]}


{
    'Matriz': [12000, 15000, 11000],
    'Filial Sul': [8000, 7200, 6500],
    'Filial Norte': [9000, 8800]
}

```python
faturamento_lojas = {
    'Matriz': [12000, 15000, 11000],
    'Filial Sul': [8000, 7200, 6500],
    'Filial Norte': [9000, 8800]
}

def total_faturamento (dicionario):
    total_faturamento_por_loja = {}
    qtd_lojas = len(dicionario)
    for loja_nome, loja_vendas in dicionario.items():
        total_faturamento_por_loja[loja_nome] = sum(loja_vendas)
    total_vendas_geral = sum(total_faturamento_por_loja.values())
    media_vendas_lojas = total_vendas_geral / qtd_lojas


    return total_faturamento_por_loja, total_vendas_geral, media_vendas_lojas


total_faturamento_por_loja, total_vendas_geral, media_vendas_lojas = total_faturamento(faturamento_lojas)

for nome, venda in total_faturamento_por_loja.items():
    print(f'Loja: {nome.upper()} - Total de vendas: {venda}')

print(f'Faturamento total: {total_vendas_geral}')
print(f'Média Faturamento: {media_vendas_lojas:.2f}')
print(f'{total_faturamento_por_loja}')
```
