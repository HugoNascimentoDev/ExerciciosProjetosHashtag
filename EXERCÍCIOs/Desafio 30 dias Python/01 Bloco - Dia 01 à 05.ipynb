Dia 01:
Calcule o total e a média de vendas do dia.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: [1200, 850, 430, 2100, 980, 1500, 760]


```python
vendas = [1200, 850, 430, 2100, 980, 1500, 760]

print(vendas)

total = sum(vendas)
media = total / len(vendas)
maior_venda = max(vendas)
menor_venda = min(vendas)

print(f'O TOTAL de Vendas é: {total}.')
print(f'A MÉDIA de vendas é {media:.2f} por registro.')
print(f'A MAIOR venda foi: {maior_venda}.')
print(f'A MENOR venda foi: {menor_venda}.')
```
