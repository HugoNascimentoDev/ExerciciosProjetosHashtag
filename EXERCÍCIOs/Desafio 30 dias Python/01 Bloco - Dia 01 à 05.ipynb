Dia 01:
Calcule o total e a média de vendas do dia.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: [1200, 850, 430, 2100, 980, 1500, 760]


```python
vendas = [1200, 850, 430, 2100, 980, 1500, 760]

print(vendas)

total = sum(vendas)
media = total / len(vendas)
maior_venda = max(vendas)
menor_venda = min(vendas)

print(f'O TOTAL de Vendas é: {total}.')
print(f'A MÉDIA de vendas é {media:.2f} por registro.')
print(f'A MAIOR venda foi: {maior_venda}.')
print(f'A MENOR venda foi: {menor_venda}.')
```

Dia 2 – Condições e contagem
Enunciado:
Conte quantos produtos possuem estoque maior que 10 e quantos estão abaixo de 5.
O que deve ser feito:
- Resolver o problema proposto
- Utilizar os conceitos estudados
- Garantir clareza e organização
Base de dados: [3, 15, 8, 22, 4, 12, 30]


```python
estoque = [3, 15, 8, 22, 4, 12, 30]

maior_dez = []
menor_cinco = []
estoque_seguro = []

for e in estoque:
    if e > 10:
        maior_dez.append(e)
    elif e < 5:
        menor_cinco.append(e)
    else:
        estoque_seguro.append(e)

print(f'''Total de produtos: {len(estoque)}
Total de produtos com estoque acima de 10 unidades: {len(maior_dez)}
Estoque acima de 10 unidades: {maior_dez}
Total de produtos com estoque abaixo de 5 unidades: {len(menor_cinco)}
Estoque abaixo de 5 unidades: {menor_cinco}
Estoque seguro: {estoque_seguro}''')
```
